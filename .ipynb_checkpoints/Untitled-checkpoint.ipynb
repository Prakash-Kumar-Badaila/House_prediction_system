```python
#data cleaing
```

```python
df_cleaned = df.copy()
```

```python
df_cleaned.head()
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.shape
```

```python
df_cleaned.dtypes
```

```python
df_cleaned ['sex'].value_counts()
```

```python
df_cleaned ['sex']= df_cleaned['sex'].map({"male":0 ,"female":1})
```

```python
df_cleaned .head()
```

```python
df_cleaned['smoker'].value_counts()
```

```python
df_cleaned['smoker'] = df_cleaned['smoker'].map({"yes":1 ,"no":0})
```

```python
df_cleaned.head()
```

```python
df_cleaned.rename(columns={
    'sex' : 'is_female',
    'smoker':'is_smoker'
},inplace=True)
```
